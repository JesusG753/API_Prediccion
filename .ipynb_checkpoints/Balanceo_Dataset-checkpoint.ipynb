```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

# Cargar y preprocesar datos
dataframe = pd.read_csv('Hipertension_Arterial_Mexico_limpo.csv', sep=';')
dataframe = dataframe.replace(',', '.', regex=True)
for col in dataframe.columns:
    try:
        dataframe[col] = dataframe[col].astype(float)
    except ValueError:
        pass

# Separamos el dataframe en x y
y = dataframe['riesgo_hipertension']
x = dataframe.drop('riesgo_hipertension', axis=1)

# Dividimos en sets de entrenamiento y prueba
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=25, stratify=y)

# Definir etiquetas para la matriz de confusión
clasificadores = ['No tiene', 'Tiene riesgo']

# Configurar el RandomForestClassifier con class_weight='balanced' y max_depth=5
Peso_randomforest = RandomForestClassifier(n_estimators=80, class_weight='balanced', max_depth=5, random_state=42)

# Entrenar el algoritmo
Peso_randomforest.fit(x_train, y_train)

# Predecir en conjunto de prueba
predicciones = Peso_randomforest.predict(x_test)

# Visualizar matriz de confusión
sns.heatmap(confusion_matrix(y_test, predicciones), annot=True, fmt="d", xticklabels=clasificadores, yticklabels=clasificadores)
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

# Reporte en conjunto de prueba
print("Reporte en conjunto de prueba (class_weight='balanced', max_depth=5):")
print(classification_report(y_test, predicciones, target_names=clasificadores))

# Reporte en conjunto de entrenamiento para verificar sobreajuste
train_pred = Peso_randomforest.predict(x_train)
print("\nReporte en conjunto de entrenamiento (class_weight='balanced', max_depth=5):")
print(classification_report(y_train, train_pred, target_names=clasificadores))

# Validación cruzada
model = RandomForestClassifier(n_estimators=80, class_weight='balanced', max_depth=5, random_state=42)
scores = cross_val_score(model, x, y, cv=5, scoring='f1_macro')
print("\nF1-score macro promedio (validación cruzada, class_weight='balanced', max_depth=5):", scores.mean())
```

```python
Peso_randomforest = RandomForestClassifier(n_estimators=80, class_weight='balanced', max_depth=4, random_state=42)
Peso_randomforest.fit(x_train, y_train)
predicciones = Peso_randomforest.predict(x_test)
print("Reporte en conjunto de prueba (class_weight='balanced', max_depth=4):")
print(classification_report(y_test, predicciones, target_names=clasificadores))
train_pred = Peso_randomforest.predict(x_train)
print("\nReporte en conjunto de entrenamiento:")
print(classification_report(y_train, train_pred, target_names=clasificadores))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# Cargar y preprocesar datos
dataframe = pd.read_csv('Hipertension_Arterial_Mexico_limpo.csv', sep=';')
dataframe = dataframe.replace(',', '.', regex=True)
for col in dataframe.columns:
    try:
        dataframe[col] = dataframe[col].astype(float)
    except ValueError:
        pass

# Separamos el dataframe en x y
y = dataframe['riesgo_hipertension']
x = dataframe.drop('riesgo_hipertension', axis=1)

# Dividimos en sets de entrenamiento y prueba
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=25, stratify=y)

# Definir etiquetas
clasificadores = ['No tiene', 'Tiene riesgo']

# Entrenar el modelo final
Peso_randomforest = RandomForestClassifier(n_estimators=80, class_weight='balanced', max_depth=5, random_state=42)
Peso_randomforest.fit(x_train, y_train)

# Evaluar en prueba
predicciones = Peso_randomforest.predict(x_test)
print("Reporte en conjunto de prueba (modelo final):")
print(classification_report(y_test, predicciones, target_names=clasificadores))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from collections import Counter

# Cargar el dataset balanceado
balanced_data = pd.read_csv('Hipertension_Arterial_Mexico_balanced.csv')

# Verificar distribución para confirmar balanceo
print("Distribución del dataset balanceado:", Counter(balanced_data['riesgo_hipertension']))

# Separar características y variable objetivo
y = balanced_data['riesgo_hipertension']
x = balanced_data.drop('riesgo_hipertension', axis=1)

# Dividir en entrenamiento y prueba
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

# Entrenar el modelo
model = RandomForestClassifier(n_estimators=119, random_state=42)
model.fit(x_train, y_train)

# Evaluar en conjunto de prueba
predicciones = model.predict(x_test)
clasificadores = ['No tiene', 'Tiene riesgo']
print("Reporte de clasificación (conjunto de prueba):")
print(classification_report(y_test, predicciones, target_names=clasificadores))

# Evaluar en conjunto de entrenamiento
train_pred = model.predict(x_train)
print("\nReporte de clasificación (conjunto de entrenamiento):")
print(classification_report(y_train, train_pred, target_names=clasificadores))

# Validación cruzada
scores = cross_val_score(model, x, y, cv=5, scoring='f1_macro')
print("\nF1-score macro promedio (validación cruzada):", scores.mean())
print("Desviación estándar (validación cruzada):", scores.std())
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from collections import Counter

# Cargar el dataset balanceado
balanced_data = pd.read_csv('Hipertension_Arterial_Mexico_balanced.csv')

# Verificar distribución para confirmar balanceo
print("Distribución del dataset balanceado:", Counter(balanced_data['riesgo_hipertension']))

# Separar características y variable objetivo
y = balanced_data['riesgo_hipertension']
x = balanced_data.drop('riesgo_hipertension', axis=1)

# Dividir en entrenamiento y prueba
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

# Entrenar el modelo con max_depth=5
model = RandomForestClassifier(n_estimators=119, max_depth=5, random_state=42)
model.fit(x_train, y_train)

# Evaluar en conjunto de prueba
predicciones = model.predict(x_test)
clasificadores = ['No tiene', 'Tiene riesgo']
print("Reporte de clasificación (conjunto de prueba, max_depth=5):")
print(classification_report(y_test, predicciones, target_names=clasificadores))

# Evaluar en conjunto de entrenamiento
train_pred = model.predict(x_train)
print("\nReporte de clasificación (conjunto de entrenamiento, max_depth=5):")
print(classification_report(y_train, train_pred, target_names=clasificadores))

# Validación cruzada
scores = cross_val_score(model, x, y, cv=5, scoring='f1_macro')
print("\nF1-score macro promedio (validación cruzada, max_depth=5):", scores.mean())
print("Desviación estándar (validación cruzada):", scores.std())
```
